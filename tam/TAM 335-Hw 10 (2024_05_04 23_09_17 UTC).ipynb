```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
#Question 8.6

D = 3e-3

temp = np.array([0,20,40,60,80,100])
nu = np.array([1.787e-6, 1.004e-6, 6.58e-7, 4.745e-7, 3.65e-7, 2.94e-7])

Q = np.zeros(6)
for i in range(Q.size):
    Q[i] = 600 * nu[i] * np.pi * D
    
for i in range(Q.size):
    print(temp[i], "*C : Q [ft^3/s] =", Q[i])
```

```python
plt.plot(temp, Q)
plt.xlabel('Temp [*C]')
plt.ylabel('Q [ft^3/s]')
plt.title('Q vs Temp')
```
