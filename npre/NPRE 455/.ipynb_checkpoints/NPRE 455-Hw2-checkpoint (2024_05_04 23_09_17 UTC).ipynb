```python
import numpy.linalg as la
import numpy as np
from sympy import *
import matplotlib.pyplot as plt
import pandas as pd

#plt.savefig(r'C:\Users\reape\Desktop\Sp24\NPRE 455\HW2', dpi=600, edgecolor='auto')
```

```python
#1
phi_0, R, r, sigma_a, sigma_f = S('phi_0, R, r, Sigma_a, Sigma_f')

phi_0 = 13.1e13 #1/(cm^2*s)
R = 50 #cm
sigma_a = 0.108 #1/cm
sigma_f = 0.0727 #1/cm

phi = phi_0 * R * sin(pi * r / R) / r
phi
```

```python
#1b
radius = np.arange(51)
flux = np.zeros(len(radius))

for i in range(len(radius)):
    flux[i] = phi.subs(r,i).evalf()

plt.figure(facecolor='White')
plt.plot(radius, flux)
plt.title('#1b, Flux vs Radius', fontsize=14)
plt.xlabel('Radius [cm]', fontsize=12)
plt.ylabel('Flux [1e14 neutrons/cm^2/s]', fontsize=12)
plt.yscale('log')
```

```python
#1d
print('flux @ .62cm:', phi.subs(r,.62).evalf()*1e-14, 'e14')
```

```python
#4a
flux_spec4 = np.array(pd.read_csv(r"C:\Users\reape\Desktop\Sp24\NPRE 455\HW2\flux_spectrum1.csv"))

e_bin_l4 = flux_spec4[:,0]
e_bin_u4 = flux_spec4[:,1]
flux4 = flux_spec4[:,2]

bin_width4 = e_bin_u4 - e_bin_l4

#i
plt.figure(facecolor='White')
plt.plot(e_bin_l4, flux4)
plt.xscale('log')
plt.yscale('log')
plt.title('i) Flux vs Energy', fontsize=14)
plt.xlabel(r'Energy ${(eV)}$', fontsize=12)
plt.ylabel(r'Flux, $\mathregular{\phi}$ $(cm^{-2}s^{-1})$', fontsize=12)
plt.show()

#ii
plt.figure(facecolor='White')
plt.plot(e_bin_l4, flux4 / bin_width4)
plt.xscale('log')
plt.yscale('log')
plt.title('ii) Flux Normalized by Bin Width vs Energy', fontsize=14)
plt.xlabel(r'Energy ${(eV)}$', fontsize=12)
plt.ylabel(r'Flux, $\mathregular{\phi}$ Normalized by Bin Width $\mathregular{\phi}$ $(cm^{-2}s^{-1})$', fontsize=12)
plt.show()

#iii
plt.figure(facecolor='White')
unit_lethargy4= e_bin_u4 / e_bin_l4
log_leth4 = np.zeros(len(unit_lethargy4))
for i in range(len(unit_lethargy4)):
    log_leth4[i] = log(unit_lethargy4[i])

plt.plot(e_bin_l4, flux4 / log_leth)
plt.xscale('log')
plt.yscale('log')
plt.title('iii) Flux vs Energy', fontsize=14)
plt.xlabel(r'Energy ${(eV)}$', fontsize=12)
plt.ylabel(r'Flux, $\mathregular{\phi}$ Normalized by Unit Lethargy $\mathregular{\phi}$ $(cm^{-2}s^{-1})$', fontsize=12)
plt.show()
```
