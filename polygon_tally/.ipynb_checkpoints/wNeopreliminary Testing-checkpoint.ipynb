```python
#Zernikes
def gen_rhophi(gas):
    rho = np.linspace(0,1,gas)
    phi = np.linspace(0,2*pi,gas)
    rho,phi = np.meshgrid(rho,phi)
    return rho,phi

def m_checker(n,m):
    valid_m = np.arange(-n,n+1,2)
    if (m in valid_m) == False:
        raise ValueError('choosen m value is invalid. examples of valid: m = -n, -n+2, -n+4, ..., n')
    return

def n_mn(n,m):
    if m != 0:
        toRet = (2*(n+1))**(1/2)
    else:
        toRet = ((2*(n+1))/(2))**(1/2)
    return toRet

def r_mn(n,m,rho):
    stop = int((n-abs(m))/2)
    toRet = 0
    for k in range(stop+1):
        top = (-1)**k * fact(n-k) * rho**(n-2*k)
        bot = fact(k) * fact(int((n+abs(m))/2 - k)) * fact(int((n-abs(m))/2 - k))
        toRet += top/bot
    return toRet

def zernike_m(n,m,rho=gen_rhophi(gas)[0],phi=gen_rhophi(gas)[1]):
    m_checker(n,m)
    if m >= 0:
        toRet = n_mn(n,m) * r_mn(n,m,rho) * cos(m*phi)
    else:
        toRet = -n_mn(n,m) * r_mn(n,m,rho) * sin(m*phi)
    return toRet

def zernike(n,rho=gen_rhophi(gas)[0],phi=gen_rhophi(gas)[1]):
    ms = np.arange(-n,n+1,2)
    toRet = 0
    for m in ms:
        toRet += zernike_m(n,m)
    return toRet
```

```python
gas = 1000 #global array size
R0 = 5
p = 8
alpha = pi/p

rho,phi = gen_rhophi(1000)
glip = {'rho':rho, 'phi':phi}
```

```python
#Testing Orthonormality of Zernike Functions
rho, phi = gen_rhophi(1000)

drho = rho[1,1]
dphi = phi[1,1]

a = zernike_m(5,3,rho,phi)
b = zernike_m(1,-1,rho,phi)

c = np.sum(a * b * rho * drho * dphi / pi)
d = np.sum(a * a * rho * drho * dphi / pi)

print('Integral of two different Zernikes', c)
print('Integral of the same Zernike:', d)
#Both agree with expected for various n and m
```

```python
#Getting R_alpha

rho, phi = gen_rhophi(8)
#phi = np.linspace(0,2*pi,10)

def u_alpha(phi):
    x = (phi + alpha)/(2*alpha)
    x = x.astype(int)
    u = phi - x*(2*alpha)
    return u

def r_alpha(phi):
    u = u_alpha(phi)
    top = R0 * cos(alpha)
    bot = cos(u)
    return top/bot

r_alpha(phi)
```

```python
#Getting r, theta

def gen_rtheta(rho,phi):
    R = r_alpha(phi)
    r = rho*R
    theta = phi
    return r,theta

def gen_xy(rho,phi):
    R = r_alpha(phi)
    r,theta = gen_rtheta(rho,phi)
    x = r * cos(theta)
    y = r * sin(theta)
    return x,y

rho,phi = gen_rhophi(100)

x,y = gen_xy(rho,phi)
f = x*0

fig,ax = plt.subplots()
ax.contourf(x,y,f)
```

```python
def k_m(n,m,rho,phi):
    r,theta = gen_rtheta(rho,phi)
    R = r_alpha(phi)
    Z = zernike_m(n,m,rho,phi)
    K = Z
    
    dr = r[:,1]
    _,dr = np.meshgrid(dr,dr)
    dtheta = theta[1,1]
    dmu = r * dr * dtheta / pi / R**2
    
    a = np.sum(K**2 * dmu)
#    K /= a**(1/2)
#    b = np.sum(K**2 * dmu)
    return a

k_m(1,1,**glip)
```

```python
def input_function(x,y):
    f = 2*x**2 - y**2 + x**2*y - 4*x*y**2 + 5*x*y - 3*x + 5*y
    return f

def c_m(n,m,rho,phi):
    r,theta = gen_rtheta(rho,phi)
    R = r_alpha(phi)
    K = k_m(n,m,rho,phi)

    x = r * cos(theta)
    y = r * sin(theta)
    F = input_function(x,y)
    
    dr = r[:,1]
    _,dr = np.meshgrid(dr,dr)
    dtheta = theta[1,1]

    dmu = r * dr * dtheta / pi / R**2

    c = np.sum(K * F * dmu)
    return c

rho,phi = gen_rhophi(10)
c_m(1,1,rho,phi)
#agrees with previous iterations, but those are wrong so it means nothing
```

```python
def f_m(n,m,rho,phi):
    K = k_m(n,m,rho,phi)
    c = c_m(n,m,rho,phi)
    return c*K

rho,phi = gen_rhophi(1000)
x,y = gen_xy(rho,phi)
F = f_m(1,-1,rho,phi)

fix,ax = plt.subplots()
ax.contourf(x,y,F)
```

```python
def f_n(n,rho,phi):
    ms = np.arange(-n,n+1,2)
    toRet = 0
    
    for m in ms:
        toRet += f_m(n,m,rho,phi)

    return toRet

rho,phi = gen_rhophi(1000)
x,y = gen_xy(rho,phi)
F = f_n(1,rho,phi)

fix,ax = plt.subplots()
ax.contourf(x,y,F)
```

```python
def f_total(n,rho,phi):
    ns = np.arange(0,n+1,1)
    toRet = 0

    for ni in ns:
        toRet += f_n(ni,rho,phi)
    
    return toRet

rho,phi = gen_rhophi(1000)
x,y = gen_xy(rho,phi)
F = f_total(10,rho,phi)

fig, ax = plt.subplots()
plot = ax.contourf(x,y,F)
cbar = fig.colorbar(plot)
```

```python
def plot_fs(n,rho,phi):
    ns = np.arange(0,n+1,1)
    toRet = 0

    x,y = gen_xy(rho,phi)
    
    for ni in ns:
        toRet += f_n(ni,rho,phi)

        fig, ax = plt.subplots()
        ax.set_title(ni)
        plot = ax.contourf(x,y,toRet)
        cbar = fig.colorbar(plot)
        plt.show()
        
    return

def plot_analytical(size):
    rho,phi = gen_rhophi(size)
    x,y = gen_xy(rho,phi)
    F = input_function(x,y)

    fig, ax = plt.subplots()
    ax.set_title('Analytical')
    plot = ax.contourf(x,y,F)
    cbar = fig.colorbar(plot)
    plt.show()
```
