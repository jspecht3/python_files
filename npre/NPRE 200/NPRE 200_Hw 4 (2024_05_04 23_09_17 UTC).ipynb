```python
from sympy import *
import numpy as np
import numpy.linalg as la
import matplotlib.pyplot as plt
```

```python
#Radioactive Decay Numerical Solution

#input
half_life = 4.8
N_initial = 100
end_time = 24
time_step = 1

#derived perameters
decay_constant = np.log(2) / half_life
number_of_time_steps = int(end_time / time_step)

#initialize soultion variables
amount_of_atoms = np.zeros(number_of_time_steps + 1)
amount_of_atoms[0] = N_initial
time = np.arange(0, end_time + .01, time_step)

#solution variables
for i in range(number_of_time_steps):
    amount_of_atoms[i+1] = -decay_constant * amount_of_atoms[i] * time_step + amount_of_atoms[i]
    
#output
print(amount_of_atoms.size, time.size)
```

```python
#Analytic Solution

x = np.linspace(0,24,1000)

N_final_time = N_initial * np.exp(-decay_constant * end_time)
N_any_time = N_initial * np.exp(-decay_constant * x)
print(N_final_time)
```

```python
#Question 2
fig, ax = plt.subplots()

#Add this to top plot if you want dots - , marker='o', markersize=4,
ax.plot(time, amount_of_atoms, 'r', marker='o', markersize=2, label='Numerical')
ax.plot(x, N_any_time, color='blue', label='Analytical')

leg = ax.legend(fancybox=True, framealpha=.80, shadow=True, borderpad=1);

plt.xlabel('Time Elapsed (Hours)')
plt.ylabel('Number of Atoms')

plt.title('Radioactive Decay, Numerical (time step=' + str(time_step) + ') vs Analytic Method')
plt.figtext(0, -.25, "This is a graph showing the difference between the Numerical (forward difference method) \nand the Analytical Approach. For both approaches, the number of starting atoms was 100, \nthe half life was 4.8 hours, and the time elapsed was 24 hours. For the numerical approach, \nthe timestep was " + str(time_step) + "hour(s). The final value for the Numerical approach was " + str(round(amount_of_atoms[-1], 6)) + " atoms \nand for the Analytical approach, the final value was 3.125 atoms.\nPercent error = " + str(round(100 *(3.125-amount_of_atoms[-1])/(3.125+amount_of_atoms[-1]), 2)) + "%")
```
