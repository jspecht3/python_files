# Functions

```python
#Part E
def M_gen(N, d1, d2, er1, er2):
    x = np.linspace(0,a+b,N+1)
    M = np.zeros([N+1, N+1])
    h = x[1]

    #Constant Arrays
    D = np.zeros(N+1)
    E = np.zeros(N+1)
    
    for i in range(len(x)):
        if x[i] <= a:
            D[i] = d1 #D1
            E[i] = er1 #SigmaA1
        else:
            D[i] = d2 #D2
            E[i] = er2 #SigmaA2
    
    #Boundary Conditions
    np.put(M, (0,1,2), (-3,4,-1))
    M[-1,-1] = 1
    
    #Array Population
    for i in range(1, N):
        M[i,i-1] = -2*D[i]/h**2*D[i-1]/(D[i-1]+D[i]) #Lower Diag
        M[i,i] = 2*D[i]/h**2*(D[i-1]/(D[i-1]+D[i]) + D[i+1]/(D[i+1]+D[i])) + E[i] #On diag
        M[i,i+1] = -2*D[i]/h**2*D[i+1]/(D[i+1]+D[i]) #Upper Diag
    return M

def Scat_gen(N, s1, s2):
    x = np.linspace(0,a+b,N+1)
    E = np.zeros((N+1, N+1))
    for i in range(1, N):
        if x[i] <= a:
            E[i,i] = s1
        else:
            E[i,i] = s2
    return -E

def F_gen(N, ef):
    x = np.linspace(0,a+b,N+1)
    F = np.zeros((N+1,N+1))
    for i in range(1, N):
        if x[i] <= a:
            F[i,i] = ef
    return F

def MF_gen(N):
    #Getting Submatricies
    m_ul = M_gen(N, D1C, D1M, Es12C+Ea1C, Es12M+Ea1M)
    m_lr = M_gen(N, D2C, D2M, Ea2C, Ea2M)
    m_ll = Scat_gen(N, Es12C, Es12M)
    
    f_ul = F_gen(N, nuEf1)
    f_ur = F_gen(N, nuEf2)
    
    #Putting Values into M
    M = np.zeros((2*N+2, 2*N+2))
    M[0:N+1,0:N+1] = m_ul
    M[N+1:,N+1:] = m_lr
    M[N+1:,0:N+1] = m_ll
    
    #Putting Values into F
    F = np.zeros((2*N+2,2*N+2))
    F[0:N+1,0:N+1] = f_ul
    F[0:N+1,N+1:] = f_ur
    
    return M, F

def power_iteration(N):
    #Initializing
    M, F = MF_gen(N)
    x = np.concatenate((np.linspace(0,a+b,N+1),np.linspace(0,a+b,N+1)))
    
    #Guesses
    k0, phi0 = 1, np.ones(2*N+2)
    err, e1, e2 = 1e-6, 1, 1

    #Looping
    while err <= e1 or err <= e2:
        phi1 = la.inv(M)/k0@F@phi0
        k1 = np.trapz(F@phi1,x)/np.trapz(M@phi1,x)
        e1 = la.norm(phi1-phi0,2)/la.norm(phi0,2)
        e2 = la.norm(k1-k0)/la.norm(k0)
        phi0, k0 = phi1, k1
    F[0,0], F[0,N+1] = nuEf1, nuEf2

    return phi0/np.trapz(F@phi0, x), k0

#Part F
def fission_rate_diffusion():
    N = 1000
    x = np.linspace(0,a+b,N+1)
    F = MF_gen(N)[1]
    F[0,0], F[0,N+1] = nuEf1, nuEf2
    phi = F@power_iteration(N)[0]
    plt.plot(x, phi[0:N+1]/2.4, label='Diffusion')

def fission_rate_transport():
    phi = np.append(D['fast_flux_a25'][:,1], D['thermal_flux_a25'][:,1])
    x1, x2 = D['fast_flux_a25'][:,0], D['thermal_flux_a25'][:,0]
    F1, F2, = np.zeros((len(x1), len(x1))), np.zeros((len(x2), len(x2)))
    F = np.zeros((2*len(x1), 2*len(x2)))

    for i in range(len(x1)):
        if x1[i] < 25:
            F1[i][i] = nuEf1
        if x2[i] < 25:
            F2[i][i] = nuEf2

    F[0:len(x1),0:len(x1)], F[len(x1):,len(x1):] = F1, F2

    plt.plot(x1, ((F@phi)[len(x1):] + (F@phi)[0:len(x1)])/2.4, label='Transport', linestyle='--', color='r')
    return
```
