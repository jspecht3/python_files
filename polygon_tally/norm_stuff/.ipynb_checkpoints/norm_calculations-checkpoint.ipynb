```python
from glip_glorp import *
```

```python
# parameters
p,R0 = 8,5

rho,phi = gen_rhophi(1000)
r,theta = gen_rtheta(rho,phi,p,R0)
x,y = gen_xy(rho,phi,p,R0)
```

# K

## Norm Calculation for K

```python
# K basis vectors, norm and runtime

# higher order Results in dtype to be set as 0 and not float64
Fa = base_input(x,y)
Fk = 0

l2_errs_k = []
linf_errs_k = []
runtimes_k = []

ti = time.time()
for n in range(21):
    print('working on n={} calculation'.format(n))
    Fk += fk_n(n,r,theta,p,R0)
    Fd = Fa - Fk
    
    l2_err = la.norm(Fd,2) / la.norm(Fa,2)
    l2_errs_k.append(l2_err)

    linf_err = la.norm(Fd,np.inf)
    linf_errs_k.append(linf_err)

    tf = time.time()
    runtimes_k.append(tf-ti)
```

## Plotting K Errors

```python
k_twinx_dict = {
    "data1": l2_errs_k,
    "data2": linf_errs_k,
    "title": "K, L$^2$ and L$^{\infty}$ Norms",
    "xlabel": "Approximaion Order",
    "ylabel1": "L$^2$ Norm",
    "ylabel2": "L$^{\infty}$ Norm",
    "go_log_mode": 1
}

plot_twinx(**k_twinx_dict)
```

# H

## Norm Calculation for H

```python
# H basis vectors, norm and runtime

# higher order Results in dtype to be set as 0 and not float64
Fa = base_input(x,y)
Fh = 0

l2_errs_h = []
linf_errs_h = []
runtimes_h = []

ti = time.time()
for n in range(21):
    print('working on n={} calculation'.format(n))
    Fh += fh_n(n,r,theta,p,R0)
    Fd = Fa - Fh
    
    l2_err = la.norm(Fd,2) / la.norm(Fa,2)
    l2_errs_h.append(l2_err)

    linf_err = la.norm(Fd,np.inf)
    linf_errs_h.append(linf_err)

    tf = time.time()
    runtimes_h.append(tf-ti)
```

## Plotting H Errors

```python
h_twinx_dict = {
    "data1": l2_errs_h,
    "data2": linf_errs_h,
    "title": "H, L$^2$ and L$^{\infty}$ Norms",
    "xlabel": "Approximaion Order",
    "ylabel1": "L$^2$ Norm",
    "ylabel2": "L$^{\infty}$ Norm",
    "go_log_mode": 1
}

plot_twinx(**h_twinx_dict)
```

# Diff

## taking the difference

```python
l2_diff = abs(np.array(l2_errs_k) - np.array(l2_errs_h))
linf_diff = abs(np.array(linf_errs_k) - np.array(linf_errs_h))
```

## plotting

```python
diff_twinx_dict = {
    "data1": l2_diff,
    "data2": linf_diff,
    "title": "Difference of L$^2$ and L$^{\infty}$ Norms b/w K and H",
    "xlabel": "Approximaion Order",
    "ylabel1": "Diff in L$^2$ Norm",
    "ylabel2": "Diff in L$^{\infty}$ Norm",
    "go_log_mode": 1
}

plot_twinx(**diff_twinx_dict)
```
