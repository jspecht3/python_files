```python
import numpy as np
import numpy.linalg as la
from math import factorial as fac

from matplotlib import pyplot as plt
font = {'fontname':'Times New Roman'}
size = {'fontsize':'14'}
#save_path = r'C:\Users\reape\Desktop\Sp24\NPRE 455\CP1\Plots\\'
```

```python
#Part B

#Solving For Coefficients
A = np.array([(1,1,1,1),
              (0,1,2,3),
              (0,1/2,4/2,9/2),
              (0,1/6,8/6,27/6)])

B = np.transpose(np.array([(0,0,1,0)]))
C = la.inv(A)@B

[C0,C1,C2,C3] = C[:]

print('Constants Array:')
display(C)

#Checking 4th order
print('4th order Evaluation:', ((C1+(C2*2**4)+(C3*3**4))/fac(4))[0])
```

```python
#Part E

#Defining the Numerical Plotting
def dsindx2_approx(h, graph_color):
    x = np.arange(0,2*np.pi+h,h)
    y = np.zeros(len(x))
    for i in range(len(x)):
        y[i] = (C0*np.sin(x[i]) + C1*np.sin(x[i]+h) + C2*np.sin(x[i]+2*h) + C3*np.sin(x[i]+3*h))/(h**2)
    plt.plot(x, y, label = 'h = {}'.format(h), color = graph_color)

#Generating Analytical Plot of -sin(x), the second derivative of sin(x)
x_sin = np.arange(0, 2*np.pi+.1, .1)
y_sin = np.zeros(len(x_sin))
for i in range(len(x_sin)):
    y_sin[i] = -np.sin(x_sin[i])

#Plotting
h_array = np.array([1,0.5,0.25])
color_array = np.array(['navy', 'orange', 'green'])

plt.figure(dpi=1000)
for i in range(len(h_array)):
    dsindx2_approx(h_array[i], color_array[i])
plt.plot(x_sin, y_sin, label = 'Analytical', color = 'red')

plt.legend()
plt.xlabel('X value',**font, **size)
plt.ylabel('Y value', **font, **size)

plt.axvline(np.pi, linestyle='--', color = 'silver')
plt.xlim(0, 2*np.pi)

#plt.savefig(save_path+'1e')
plt.show()
```
