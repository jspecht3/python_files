```python
import numpy as np
from scipy.integrate import quad
from sympy import *

from matplotlib import pyplot as plt
import matplotlib as mpl
font = {'family' : 'Times New Roman',
        'weight' : 'normal',
        'size'   : 12}
mpl.rc('font', **font)

import warnings
warnings.filterwarnings('ignore')

np.set_printoptions(edgeitems=30, linewidth=100000, 
    formatter=dict(float=lambda x: "%.5g" % x))

save = r'C:\Users\reape\Desktop\Sp24\NPRE 421\Hw 8\\'
```

```python
#Constants
rxn = ['DT','DHe3','DD_pT','DD_nHe3']
xs = []
E = np.linspace(10,1e4,100000)

A = np.array([[45.95,50200,1.37e-2,1.076,409],
              [89.27,25900,3.98e-3,1.297,647],
              [46.097,372,4.36e-4,1.22,0],
              [47.88,482,3.08e-4,1.177,0]])

for i in range(len(rxn)):
    a1, a2, a3, a4, a5 = A[i]
    top = a5 + (a2/((a4-a3*E)**2+1))
    bot = E*(np.exp(a1/E**(.5))-1)
    xs.append(top/bot)

for i in range(len(rxn)-2):
    plt.loglog(E, xs[i], label=rxn[i])
plt.plot(E, xs[2]+xs[3], label='DD')
plt.title('1a, All')
plt.legend(), plt.grid(which='both')
plt.xlabel('Energy [keV]'), plt.ylabel('Cross Section [b]')
plt.savefig(save+'1a_total', dpi=600)
plt.show()

plt.semilogx(E, xs[0], label=rxn[1])
plt.legend(), plt.grid(which='both')
plt.title('1a, DT')
plt.xlabel('Energy [keV]'), plt.ylabel('Cross Section [b]')
plt.savefig(save+'1a_DT', dpi=600)
plt.show()

plt.semilogx(E, xs[1], label=rxn[1])
plt.legend(), plt.grid(which='both')
plt.title('1a, D-He3')
plt.xlabel('Energy [keV]'), plt.ylabel('Cross Section [b]')
plt.savefig(save+'1a_DHe3', dpi=600)
plt.show()

plt.semilogx(E, xs[2], label=rxn[2])
plt.semilogx(E, xs[3], label=rxn[3])
plt.semilogx(E, xs[2]+xs[3], label='DD')
plt.legend(), plt.grid(which='both')
plt.title('1a, DD')
plt.xlabel('Energy [keV]'), plt.ylabel('Cross Section [b]')
plt.savefig(save+'1a_DD')
plt.show()
```

```python
#Part C

def dt_xs(E):
    a1, a2, a3, a4, a5 = A[0]
    top = a5 + (a2/((a4-a3*E)**2+1))
    bot = E * (np.exp(a1/E**(.5))-1)
    return top/bot

#Constants
md = 3.34358344e-27 #kg
mt = 5.00808509e-27 #kg
mr = md*mt/(md+mt)
kb = 1 #8.617e-5 #KeV/K

T_space = np.linspace(10,1e4,1000)
rc_arr = []

def integrand(E):
    return dt_xs(E) * E * np.exp(-E/kb/T)

for T in T_space:
    C = (8/np.pi)**(1/2) * (mr/kb/T)**(3/2) / mr**2
    rc = C * quad(integrand, 0, np.inf)[0]
    rc_arr.append(rc)

plt.loglog(T_space, rc_arr)
plt.title('1C, Rate Coefficient'), plt.grid(which='both')
plt.xlabel('Ion Energy [keV]'), plt.ylabel('Rate Coefficient [cm$^{3}$/s]')
plt.savefig(save+'1c')
plt.show()
```
