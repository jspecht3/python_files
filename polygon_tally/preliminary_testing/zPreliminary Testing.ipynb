```python
#Zernike Functions#############################################################################################################
def m_checker(n,m):
    valid_m = np.arange(-n,n+1,2)
    if (m in valid_m) == False:
        raise ValueError('choosen m value is invalid. examples of valid: m = -n, -n+2, -n+4, ..., n')
    return

def R_nm(n,m,rho): #CHECKED AND GOOD
    stop = int((n-abs(m))/2)
    to_ret = 0
    for k in range(stop+1):
        top = (-1)**k * math.factorial(n-k) * math.factorial(k) * rho**(n-2*k)
        bot = math.factorial(int(((n+abs(m))/2)) - k) * math.factorial(int(((n-abs(m)))/2))
        to_ret += top/bot
    return to_ret

def N_nm(n,m): #CHECKED AND GOOD
    to_ret = 0
    if m==0:
        return ((2*(n+1))/(1+1))**(1/2)
    else:
        return ((2*(n+1))/(1))**(1/2)

def zernike(n): #CHECKED AND GOOD
    rho, phi = rho_phi_gen()
    m = np.arange(-n,n+1,2)
    toRet = 0
    
    for j in m:
        if j >= 0:
            toRet += N_nm(n,j) * R_nm(n,j,rho) * np.cos(j*phi)
        else:
            toRet += -N_nm(n,j) * R_nm(n,j,rho) * np.sin(j*phi)
    return toRet

def zernike_m(n,m):
    m_checker(n,m)
    
    rho, phi = rho_phi_gen()
    toRet = 0
    
    if m >= 0:
        toRet += N_nm(n,m) * R_nm(n,m,rho) * np.cos(m*phi)
    else:
        toRet += -N_nm(n,m) * R_nm(n,m,rho) * np.sin(m*phi)
    return toRet
```

```python
#Plotting Functions#######################################################################################################
def plot_polygon(p,R0):
    x,y = polygon_gen_cart(p,R0)
    f = 0*x
    width = 1.1*R0

    fig, ax = plt.subplots(figsize=(5,5))
    ax.set_xlim(-width,width)
    ax.set_ylim(-width,width)
    ax.set_xlabel('X position')
    ax.set_ylabel('Y position')
    ax.set_title('{}-Sided Regular Polygon w/ Radius={}'.format(p,R0))
    ax.contourf(x,y,f)
    plt.show()

def plotter(x,y,z):
    fig, ax = plt.subplots()
    ax.contourf(x,y,z)
    plt.show()
```

```python
#Basic Polygon Functions######################################################################################################
def rho_phi_gen():
    rho = np.linspace(0,1,global_array_size)
    phi = np.linspace(0,2*pi,global_array_size)
    rho, phi = np.meshgrid(rho,phi)
    return rho, phi

def uv_gen():
    rho, phi = rho_phi_gen()
    u = rho*cos(phi)
    v = rho*sin(phi)
    return u,v

def r_alpha(p,R0):
    alpha = pi/p
    _, phi = rho_phi_gen()
    
    x = (phi + alpha)/(2*alpha)
    U = np.zeros_like(phi)

    for i in range(len(phi)):
        U[i] = phi[i,i] - int(x[i,i])*(2*alpha)
    
    return R0*cos(alpha)/cos(U)

def polygon_gen_polar(p,R0):
    alpha = pi/p
    
    rho, phi = rho_phi_gen()
    R = r_alpha(p,R0) 
    
    r = rho*R
    theta = phi

    return r, theta

def polygon_gen_cart(p,R0):
    r, theta = polygon_gen_polar(p,R0)
    
    x = r*cos(theta)
    y = r*sin(theta)

    return x,y
```

```python
#Working Functions
def checker(n,p,R0):
    x,y = polygon_gen_cart(p,R0)

    f1 = input_function(x,y)
    f2 = f_cumulative(n,p,R0)

    print('Analytical Solution')
    plotter(x,y,f1)
    print('Zernike Expansion, Order={}'.format(n))
    plotter(x,y,f2)

def zernikes_all_m(n,p,R0):
    x,y = polygon_gen_cart(8,5)

    ms = np.arange(-n,n+1,2)

    for m in ms:
        f = zernike_m(n,m)
        plotter(x,y,f)

#F constituents##################################################################################################
def input_function(x,y):
    return 2*x**2 - y**2 + x**2*y - 4*x*y**2 + 5*x*y - 3*x + 5*y

def k_polar(n,p,R0):
    r, theta = polygon_gen_polar(p,R0)
    toRet = zernike(n)*(r/r_alpha(p,R0))
    return toRet

def k_polar_m(n,m,p,R0):
    r, theta = polygon_gen_polar(p,R0)
    toRet = zernike_m(n,m)*(r/r_alpha(p,R0))
    return toRet

######TO DEL####################################
def c_testing(n,m,p,R0):
    r,theta = polygon_gen_polar(p,R0)

    dr = r[:,1]
    dtheta = theta[1,:]
    dr,dtheta = np.meshgrid(dr,dtheta)
    
    x = r*cos(theta)
    y = r*sin(theta)

    f = input_function(x,y)
    R = r_alpha(p,R0)
    K = k_polar_m(n,m,p,R0)

    toRet = 1/pi * f * K * r * dr * dtheta / R**2
    toRet = np.sum(toRet)
    
    return toRet, r, theta, dr, dtheta
################################################

#XXXXXXXXXXXXXXXXXXXXXXXXXXXXXXXXXXXXXXXXXXXXXXXXXXXXX
#Undoctored cmn
def c_mn(n,m,p,R0):
    r,theta = polygon_gen_polar(p,R0)

    dr = r[:,1]
    dtheta = theta[1,:]
    dr,dtheta = np.meshgrid(dr,dtheta)
    
    x = r*cos(theta)
    y = r*sin(theta)

    f = input_function(x,y)
    R = r_alpha(p,R0)
    K = k_polar_m(n,m,p,R0)

    toRet = 1/pi * f * K * r * dr * dtheta / R**2
    toRet = np.sum(toRet)
    
    return toRet
#XXXXXXXXXXXXXXXXXXXXXXXXXXXXXXXXXXXXXXXXXXXXXXXXXXXXXXX

def f_nm(n,m,p,R0):
    m_checker(n,m)
    
    c = c_mn(n,m,p,R0)
    K = k_polar_m(n,m,p,R0)

    return c*K

def f_n(n,p,R0):
    ms = np.arange(-n,n+1,2)
    toRet = 0

    for m in ms:
        toRet += f_nm(n,m,p,R0)
    return toRet

def f_cumulative(n,p,R0):
    ns = np.arange(n+1)
    toRet = 0
    
    for n in ns:
        toRet += f_n(n,p,R0)
    return toRet


#Nincompoop functions######################################################################################################
def uv_from_xy(x,y,p,R0):
    R = r_alpha(p,R0)

    u = x/R/atan(y/x)
    v = y/R/atan(y/x)

    return u,v
```

```python
#Koch Snowflake
gleep = 7

def koch_snowflake(order, scale=10):
    def _koch_snowflake_complex(order):
        if order == 0:
            # initial triangle
            angles = np.array([0, 120, 240]) + 90
            return scale / np.sqrt(3) * np.exp(np.deg2rad(angles) * 1j)
        else:
            ZR = 0.5 - 0.5j * np.sqrt(3) / 3

            p1 = _koch_snowflake_complex(order - 1)  # start points
            p2 = np.roll(p1, shift=-1)  # end points
            dp = p2 - p1  # connection vectors

            new_points = np.empty(len(p1) * 4, dtype=np.complex128)
            new_points[::4] = p1
            new_points[1::4] = p1 + dp / 3
            new_points[2::4] = p1 + dp * ZR
            new_points[3::4] = p1 + dp / 3 * 2
            return new_points

    points = _koch_snowflake_complex(order)
    x, y = points.real, points.imag
    return x, y

x,y = koch_snowflake(gleep)
plt.figure(figsize=(8, 8))
plt.axis('equal')
plt.fill(x, y)
plt.show()
```
