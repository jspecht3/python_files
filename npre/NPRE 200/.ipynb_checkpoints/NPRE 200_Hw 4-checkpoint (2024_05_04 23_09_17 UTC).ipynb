```python
from sympy import *
import numpy as np
import numpy.linalg as la
import matplotlib.pyplot as plt
```

```python
#Radioactive Decay Analytic Solution

#input
half_life = 4.8
N_initial = 100
end_time = 24
time_step = 1

#derived perameters
decay_constant = np.log(2) / half_life
number_of_time_steps = int(end_time / time_step)

#initialize soultion variables
amount_of_atoms = np.zeros(number_of_time_steps + 1)
amount_of_atoms[0] = N_initial
time = np.arange(0, number_of_time_steps + 1, time_step)

#solution variables
for i in range(number_of_time_steps):
    amount_of_atoms[i+1] = -decay_constant * amount_of_atoms[i] * time_step + amount_of_atoms[i]
    
#output
print(amount_of_atoms[24], time[24])
```

```python
#Numerarical Solution

x = np.linspace(0,24,1000)

N_final_time = N_initial * np.exp(-decay_constant * end_time)
N_any_time = N_initial * np.exp(-decay_constant * x)
print(N_final_time)
```

```python
fig, ax = plt.subplots()

ax.plot(time, amount_of_atoms, 'r', marker='o', markersize=4, label='Numerical')
ax.plot(x, N_any_time, color='blue', label='Analytical')

leg = ax.legend(fancybox=True, framealpha=.80, shadow=True, borderpad=1);

plt.xlabel('Time Elapsed (Hours)')
plt.ylabel('Number of Atoms')

plt.title('Radioactive Decay of Numerical vs Analytical Approach')

plt.figtext(0, -.2, "This is a graph showing the difference between the Numerical (forward difference method) \nand the Analytical Approach. For both approaches, the number of starting atoms was 100, \nthe half life was 4.8 hours, and the time elapsed was 24 hours. For the numerical approach, \nthe timestep was 1 hour. The final value for the Numerical approach was 2.368261 atoms \nand for the Analytical approach, the final value was 3.125 atoms.")

#plot(x, y, color='green', marker='o', linestyle='dashed',
#    linewidth=2, markersize=12)
```
