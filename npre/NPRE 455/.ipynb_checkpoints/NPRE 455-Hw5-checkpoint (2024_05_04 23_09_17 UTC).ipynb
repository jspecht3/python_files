```python
import numpy as np
import numpy.linalg as la
from matplotlib import pyplot as plt
from tabulate import tabulate
from sympy import *
from sympy.plotting import plot
import scipy.special as sp

#Plot Variables
font = {'fontname':'Times New Roman'}
size = {'fontsize':'14'}
```

```python
#Variables
C1, C2, C3, C4 = S('C1, C2, C3, C4')
D1, D2 = S('D1, D2')
L1, L2 = S('L1, L2')
B1, B2 = S('B1, B2')
a, b, x, y, R, r = S('a, b, x, y, R, r')

SigmaF, SigmaS, SigmaA = S('Sigma_F, Sigma_S, Sigma_A')
k, nu, eta = S('k, nu, eta')
```

```python
#3
x_tilde3 = (a/2 + b) + 2*D1
Flux1_3 = sinh((x_tilde3-a/2)/L2)/cos(B1*a/2)*cos(B1*x)
Flux2_3 = sinh((x_tilde3-x)/L2)

Flux1_3s = Flux1_3.subs([(D1, 0.79), (a, 50), (B1, 0.03384), (D2, 1), (b, 50), (L2, 3.7537)]).evalf()
Flux2_3s = Flux2_3.subs([(D1, 0.79), (a, 50), (B1, 0.03384), (D2, 1), (b, 50), (L2, 3.7537)]).evalf()

max3 = Flux1_3s.subs(x,0)

FluxT_3 = np.zeros(100)
x3 = np.arange(100)
for i in range(26):
    FluxT_3[i] = Flux1_3s.subs(x,i)
for i in range(26,100):
    FluxT_3[i] = Flux2_3s.subs(x,i)

plt.ylabel('Flux [n/cm^2]',**font, **size)
plt.xlabel('x [cm]', **font, **size)
plt.title('3c', **font, **size)
plt.plot(x3, FluxT_3/max3)
plt.axvline(25, linestyle='--', label = 'Reflector Interface', color = 'r')
plt.axvline(75, linestyle='--', label = 'Outer Boundary', color = 'g')
plt.legend()
plt.show()
```

```python
#4
b_tilde4 = R + b + 2*D2
Flux1_4 = 1/r*sinh((b_tilde4-R)/L2)/sin(B1*R)*sin(B1*r)
Flux2_4 = 1/r*sinh((b_tilde4-r)/L2)

Flux1_4s = Flux1_4.subs([(D1, 0.79), (R, 25), (B1, 0.03384), (D2, 1), (b, 50), (L2, 3.7537)]).evalf()
Flux2_4s = Flux2_4.subs([(D1, 0.79), (R, 25), (B1, 0.03384), (D2, 1), (b, 50), (L2, 3.7537)]).evalf()

FluxT_4[0] = FluxT_4[1]
max4 = Flux1_4s.subs(r,1)

FluxT_4 = np.zeros(100)
r4 = np.arange(100)
for i in range(26):
    FluxT_4[i] = Flux1_4s.subs(r,i)
for i in range(26,100):
    FluxT_4[i] = Flux2_4s.subs(r,i)

plt.ylabel('Flux [n/cm^2]',**font, **size)
plt.xlabel('r [cm]', **font, **size)
plt.title('4c', **font, **size)
plt.plot(r4, FluxT_4/max4)
plt.axvline(25, linestyle='--', label = 'Reflector Interface', color = 'r')
plt.axvline(75, linestyle='--', label = 'Outer Boundary', color = 'g')
plt.legend()
plt.show()
```
