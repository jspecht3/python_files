## Convenience Functions

```python
clad = openmc.Material()
clad.add_element('Zr', 1.)
clad.set_density('g/cc', 5.)

water = openmc.Material()
water.add_nuclide('H1', 2.)
water.add_nuclide('O16', 1.)
water.set_density('g/cc', 1.)
```

### Hexagonal Prism

```python
hex_prism = openmc.model.HexagonalPrism(edge_length=2., orientation='x', corner_radius=.5)
zcyl = openmc.ZCylinder(r=1)

pin = openmc.Cell(fill=clad, region=-zcyl)
outside_pin = openmc.Cell(fill=water, region=+zcyl & -hex_prism)
```

```python
uni = openmc.Universe(cells=[pin, outside_pin])
uni.plot(width=(5,5))
```

### Isogonal Octagon

```python
stop_sign = openmc.model.IsogonalOctagon((0,0), 1.5, 1.2)
cell = openmc.Cell(region=-stop_sign, fill=None)
univ = openmc.Universe(cells=[cell])
univ.plot(width=(5,5))
```

## Random Sphere Packing

```python
h = 20
r = 10
cylinder = openmc.model.RightCircularCylinder(center_base=(0,0,-10), height=h, radius=r)
```

```python
r_sphere = 1
centers = openmc.model.pack_spheres(radius=r_sphere, region=-cylinder, pf=.1, seed=1)
```

```python
centers.shape
print(centers[:10])
```

```python
from math import pi
volume_cyl = pi * r**2 * h
volume_sphere = 4/3 * pi ** r_sphere**3

volume_sphere * 150 / volume_cyl
```

```python
triso_layer_r = [.5, .6, .7, .8]
sphere_surfs = [openmc.Sphere(r=tr) for tr in triso_layer_r]
```

```python
regions = openmc.model.subdivide(sphere_surfs)
print(regions)
```

```python
fill_mats = [water, clad, None, water, clad]
cells = []
for i in range(len(regions)):
    cells.append(openmc.Cell(fill=fill_mats[i], region=regions[i]))
```

```python
sphere_univ = openmc.Universe(cells=cells)
```

```python
sphere_univ.plot(width=(2,2))
```

```python
trisos = [openmc.model.TRISO(outer_radius=r_sphere, fill=sphere_univ, center=c) for c in centers]
print(trisos[0:3])
```

```python
graphite = openmc.Material()
graphite.add_element('C', 1.)
graphite.set_density('g/cc', 2.)
```

```python
# not in each of the trisos
region_background = openmc.Intersection([~t.region for t in trisos])
background_cell = openmc.Cell(region=region_background & -cylinder, fill=graphite)
```

```python
univ = openmc.Universe(cells=[background_cell] + trisos)
univ.plot(width=(20., 20.), pixels=[500, 500])
```

```python
lower_left, upper_right = (-cylinder).bounding_box

print(lower_left)
print(upper_right)

shape = (5,5,5)
pitch = (upper_right - lower_left)/shape
print(pitch)
```

```python
lattice = openmc.model.create_triso_lattice(trisos=trisos, lower_left=lower_left, pitch=pitch, shape=shape, background=graphite)
```

```python
cell = openmc.Cell(fill=lattice, region=-cylinder)
univ = openmc.Universe(cells=[cell])
```

```python
univ.plot(width=(20., 20.), pixels=[500, 500])
```

```python
univ.plot(width=(20., 20.), pixels=[500, 500], color_by='material')
```

## Energy Released per Fission

```python
fission_tally = openmc.Tally()
fission_tally.scores = ['kappa-fission', 'fission']
print(fission_tally)
```

```python
model.tallies = openmc.Tallies([fission_tally])
```

```python
model.settings.batches = 50
model.settings.inactive = 10
model.settings.particles = 10_000
```

```python
statepoint = model.run()
```

```python
!cat tallies.out
```

```python
with openmc.StatePoint(statepoint) as sp:
    my_tally = sp.get_tally(id=fission_tally.id)
```

```python
print(my_tally)
```

```python
# material filter, nuclide filter, score
my_tally.shape
```

```python
fission_rate = my_tally.get_values(scores=['fission']).flatten()[0]
print(fission_rate) # units of rxns per source

kappa_fission_rate = my_tally.get_values(scores=['kappa-fission']).flatten()[0]
print(kappa_fission_rate) # units of eV per source particle

kappa_fission_rate / fission_rate * 1e-6 # should be 200 MeV
```

```python
#should report the uncertainty w/ EVERY MONTE CARLO uncertainty
from uncertainties import ufloat
```

```python
fission_rate_sd = my_tally.get_values(scores=['fission'], value='std_dev').flatten()[0]
print(fission_rate_sd) # units of rxns per source

kappa_fission_rate_sd = my_tally.get_values(scores=['kappa-fission'], value='std_dev').flatten()[0]
print(kappa_fission_rate_sd) # units of eV per source particle

#creating ufloat values
f = ufloat(fission_rate, fission_rate_sd)
kf = ufloat(kappa_fission_rate, kappa_fission_rate_sd)
kf/f * 1e-6
```

```python
rel_err = 0.5358318642092597 / 193.67284658940707
print(rel_err * 100)
```

## Plot the Neutron Flux Spectrum

```python
energy_filter = openmc.EnergyFilter.from_group_structure('UKAEA-1102')
len(energy_filter.bins)
```

```python
spectrum_tally = openmc.Tally()
spectrum_tally.filters = [energy_filter]
spectrum_tally.scores = ['flux']
print(spectrum_tally)
```

```python
model.tallies += [spectrum_tally]
```

```python
statepoint = model.run()
```

```python
!cat tallies.out
```

```python
with openmc.StatePoint(statepoint) as sp:
    tally = sp.get_tally(id=spectrum_tally.id)

print(tally)
```

```python
spectrum = tally.get_values(value='mean').flatten()
print(spectrum[:5])
```

```python
import matplotlib.pyplot as plt
import numpy as np

unit_lethargy = energy_filter.lethargy_bin_width
```

```python
print(unit_lethary)
```

```python
print(np.unique(energy_filter.bins))
print(len(np.unique(energy_filter.bins)))
```

```python
plt.step(np.unique(energy_filter.bins)[:-1], spectrum/unit_lethargy)

plt.xscale('log')
plt.xlabel('Energy [eV]')
plt.ylabel('Flux per Unit Lethary')
plt.show()
```

## Normalizing Tallies

```python
lower_left, upper_right = model.geometry.bounding_box
print(lower_left)
print(upper_right)

lower_left[-1] = 0
upper_right[-1] = 1

volume = np.prod(upper_right - lower_left)
print(volume)
```

```python
with openmc.StatePoint(statepoint) as sp:
    kf_tally = sp.get_tally(id=fission_tally.id)
```

```python
kf = kf_tally.get_values(scores=['kappa-fission']).flatten()[0]
```

```python
print(kf)
```

```python
# getting the flux
J_to_eV = 1 / 1.602e-19
power = 200
neutron_source = power * J_to_eV * 1/kf
print(neutron_source*1e-13)
```

```python
normalized_spectrum = spectrum / volume * neutron_source
```

```python
plt.step(np.unique(energy_filter.bins)[:-1], normalized_spectrum/unit_lethargy)

plt.xscale('log')
plt.xlabel('Energy [eV]')
plt.ylabel('Flux per Unit Lethary [n/cm$^2$/s')
plt.show()
```
