```python
import openmc
import scipy
import matplotlib.pyplot as plt
import os

os.environ['OPENMC_CROSS_SECTIONS']
```

```python
p1 = '/home/joe/projects/endfb71_hdf5/neutron/Xe135.h5'
d1 = openmc.data.IncidentNeutron.from_hdf5(p1)
EA1 = d1[27].xs['294K'] #27 is absorption key
E1 = d1.energy['294K']

p2 = '/home/joe/projects/endfb71_hdf5/neutron/U235.h5'
d2 = openmc.data.IncidentNeutron.from_hdf5(p2)
EA2 = d2[27].xs['294K'] #27 is absorption key
E2 = d2.energy['294K']

p3 = '/home/joe/projects/endfb71_hdf5/neutron/U238.h5'
d3 = openmc.data.IncidentNeutron.from_hdf5(p3)
EA3 = d3[27].xs['294K'] #27 is absorption key
E3 = d3.energy['294K']
```

```python
#plt.plot(data)
plt.loglog(E3, EA3(E3), label='U238', linewidth=1)
plt.loglog(E1, EA1(E1), label='Xe135', linewidth=1)
plt.loglog(E2, EA2(E2), label='U235', linewidth=1)
plt.legend(), plt.xlabel('Energy [kEv]'), plt.ylabel('Neutron Absorption Cross Section [b]')
plt.show()
```
