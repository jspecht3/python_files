```python
#Functions
#Part F
def M_gen(N, d1, d2, er1, er2, Source):
    x = np.linspace(0,a+b,N+1)
    M = np.zeros([N+1, N+1])
    h = x[1]

    #Constant Arrays
    D = np.zeros(N+1)
    E = np.zeros(N+1)
    
    for i in range(len(x)):
        if x[i] <= a:
            D[i] = d1 #D1
            E[i] = er1 #SigmaA1
        else:
            D[i] = d2 #D2
            E[i] = er2 #SigmaA2
    
    S = Source * np.zeros(N+1)
    for i in range(1,len(x)-1):
        if x[i] <= a:
            S[i] = Source
    
    #Boundary Conditions
    np.put(M, (0,1,2), (-3,4,-1))
    M[-1,-1] = 1
    
    #Array Population
    for i in range(1, N):
        M[i,i-1] = -2*D[i]/h**2*D[i-1]/(D[i-1]+D[i]) #Lower Diag
        M[i,i] = 2*D[i]/h**2*(D[i-1]/(D[i-1]+D[i]) + D[i+1]/(D[i+1]+D[i])) + E[i] #On diag
        M[i,i+1] = -2*D[i]/h**2*D[i+1]/(D[i+1]+D[i]) #Upper Diag
    return M, S

def Scat_gen(N, s1, s2):
    x = np.linspace(0,a+b,N+1)
    E = np.zeros((N+1, N+1))
    for i in range(1, N):
        if x[i] <= a:
            E[i,i] = s1
        else:
            E[i,i] = s2
    return -E

def Flux_gen(N):
    #Getting Submatricies
    m_ul, f_ul = M_gen(N, D1C, DM, R1C, R1M, S)
    m_lr, f_lr = M_gen(N, D2C, D2M, R2C, R2M, S2)
    m_ll = Scat_gen(N, Es12C, Es12M)
    
    #Putting Values into M
    M = np.zeros((2*N+2, 2*N+2))
    M[0:N+1,0:N+1] = m_ul
    M[N+1:,N+1:] = m_lr
    M[N+1:,0:N+1] = m_ll
    
    #Putting Values into F
    F = np.zeros(2*N+2)
    F[0:N+1] = f_ul
    F[N+1:] = f_lr
    
    return la.inv(M)@F

#Part G
def L2_error(N):
    top1 = la.norm(np.array(Flux_gen(N)[0:N+1] - flux1(N)[1], dtype=float), 2)
    bot1 = la.norm(np.array(flux1(N)[1], dtype=float), 2)
    e1 = top1/bot1
    
    top2 = la.norm(np.array(Flux_gen(N)[N+1:] - flux2(N)[1], dtype=float), 2)
    bot2 = la.norm(np.array(flux2(N)[1], dtype=float), 2)
    e2 = top2/bot2
    return e1, e2
```
