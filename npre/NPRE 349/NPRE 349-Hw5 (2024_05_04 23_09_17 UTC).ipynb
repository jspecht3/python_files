```python
import numpy as np
import numpy.linalg as la
from sympy import *
from sympy.plotting import plot
from matplotlib import pyplot as plt

import matplotlib as mpl
mpl.rc('font', family='Times New Roman', size=12)

font = {'fontname':'Times New Roman'}
size = {'fontsize':'14'}
```

```python
#8.5b Plotting Pressure Drop and Power
D, mdot = S('D, mdot')

rho = 884.1
nu = 550e-6
L = 2.5
D = 10e-3

A = 25*np.pi*D**2/4
v = mdot/rho/A

Re = v*D/nu
f = 64/Re

deltaP = f*rho*v**2*L/2/D
P = deltaP*mdot/rho

mdot_arr = np.linspace(10, 30, 1000)
deltaP_arr = np.zeros(len(mdot_arr))
P_arr = np.zeros(len(mdot_arr))

for i in range(len(mdot_arr)):
    deltaP_arr[i] = deltaP.subs(mdot, mdot_arr[i])
    P_arr[i] = P.subs(mdot, mdot_arr[i])
    
#Plotting
print('Functions:')
display(deltaP, P)

plt.figure(dpi=600)
plt.plot(mdot_arr, deltaP_arr)
plt.title('8.5b, Pressure Drop')
plt.xlabel('Mass Flow Rate [kg/s]')
plt.ylabel('Pressure Drop [N/m/s]')
plt.show()

plt.figure(dpi=600)
plt.plot(mdot_arr, P_arr)
plt.title('8.5b, Power')
plt.xlabel('Mass Flow Rate [kg/s]')
plt.ylabel('Power [W]')
plt.show()
```
