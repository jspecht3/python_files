```python
import numpy as np
import numpy.linalg as la
from matplotlib import pyplot as plt
from tabulate import tabulate
from sympy import *
from sympy.plotting import plot
```

## Question 1

```python
x = S('x')

q = 1
m = 1
dt = .1

v0 = Matrix([1,1,1])
r0 = Matrix([0,0,0])

B = Matrix([0,0,x])
E = Matrix([0,0,0])

omega = q*B/m
sigma = q*E/m

A = omega * dt/2

v_old = v0
r_old = r0

C = v_old + dt*(sigma + v_old.cross(omega/2))

x0 = [r_old[0]]
y0 = [r_old[1]]
t = np.arange(0,5,0.1)

for i in range(len(t)):
    v_new = (C + A*(A.dot(C)) - A.cross(C)/(1 + A.dot(A))).subs(x, r_old[0])
    r_new = r_old + v_new*dt
    v_old = v_new
    r_old = r_new
    x0.append(r_new[0])
    y0.append(r_new[1])

del x0[50]
del y0[50]

plt.figure(dpi = 600)
plt.plot(t, x0, label = 'x')
plt.plot(t, y0, label = 'y')
plt.title('1. No E Field')
plt.legend()
```
